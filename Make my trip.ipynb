```python
# @title Top 10 months for query
import plotly.express as px
import pandas as pd

# Calculate the monthly frequencies
monthly_frequencies = df['QUERY DATE'].dt.month_name().value_counts().head(10)

# Convert the series to a DataFrame for Plotly
monthly_freq_df = monthly_frequencies.reset_index()
monthly_freq_df.columns = ['Month', 'Frequency']  # Rename columns

# Create the bar chart using Plotly Express
fig = px.bar(monthly_freq_df,
             x='Frequency',
             y='Month',  # Switched x and y for horizontal bar chart
             orientation='h',  # Set orientation to horizontal
             title='Top 10 Query Months',
             labels={'Frequency': 'Number of Queries'},
             color_discrete_sequence=px.colors.qualitative.Pastel)

# Customize layout
fig.update_layout(yaxis={'categoryorder': 'total ascending'})  # Sort y-axis by frequency

# Show the chart
fig.show()
```

```python
# @title Average food rating across cities
import plotly.express as px
import pandas as pd

# Assuming 'df' is your original dataframe

# 1. Get the top 10 cities based on frequency
top_10_cities = df['CITY'].value_counts().head(10).index.tolist()

# 2. Filter the dataframe for these top 10 cities
filtered_df = df[df['CITY'].isin(top_10_cities)]

# 3. Calculate average food rating for each city
city_food_ratings = filtered_df.groupby('CITY')['RATING BASED ON FOOD'].mean().reset_index(name='AVERAGE_FOOD_RATING')

# 4. Create the grouped bar chart using Plotly Express
fig = px.bar(city_food_ratings,
             x='CITY',
             y='AVERAGE_FOOD_RATING',  # Use average food rating as 'count'
             color='AVERAGE_FOOD_RATING',
             barmode='group',
             title='Average Food Ratings Across Top 10 Cities')

fig.update_layout(
    xaxis_title='City',
    yaxis_title='Average Food Rating',
    legend_title='Average Food Rating',
    xaxis={'categoryorder': 'total descending'}  # Sort x-axis by total count
)

fig.show()
```

```python
# @title Average value rating across cities
import plotly.express as px
import pandas as pd

# Assuming 'df' is your original dataframe

# 1. Get the top 10 cities based on frequency
top_10_cities = df['CITY'].value_counts().head(10).index.tolist()

# 2. Filter the dataframe for these top 10 cities
filtered_df = df[df['CITY'].isin(top_10_cities)]

# 3. Calculate average 'RATING BASED ON VALUE FOR MONEY' for each city
city_value_ratings = filtered_df.groupby('CITY')['RATING BASED ON VALUE FOR MONEY'].mean().reset_index(name='AVERAGE_VALUE_RATING')

# 4. Create the grouped bar chart using Plotly Express
fig = px.bar(city_value_ratings,
             x='CITY',
             y='AVERAGE_VALUE_RATING',
             color='AVERAGE_VALUE_RATING',
             barmode='group',
             title='Average Value Ratings Across Top 10 Cities')

fig.update_layout(
    xaxis_title='City',
    yaxis_title='Average Value Rating',
    legend_title='Average Value Rating',
    xaxis={'categoryorder': 'total descending'}  # Sort x-axis by total count
)

fig.show()
```

```python
# @title Average experience
import plotly.express as px
import pandas as pd

# Assuming 'df' is your original dataframe

# 1. Get the top 10 cities based on frequency
top_10_cities = df['CITY'].value_counts().head(10).index.tolist()

# 2. Filter the dataframe for these top 10 cities
filtered_df = df[df['CITY'].isin(top_10_cities)]

# 3. Calculate average values for the specified columns
experience_cols = ['EXCELLENT EXPERIENCE REPORTED BY VISITORS',
                   'AVERAGE EXPERIENCE REPORTED BY VISITOR',
                   'POOR EXPERIENCE REPORTED BY VISITORS',
                   'TERRIBLE EXPERIENCE REPROTED BY VISITOR']

city_experience_ratings = filtered_df.groupby('CITY')[experience_cols].mean().reset_index()

# 4. Melt the DataFrame for Plotly Express
melted_df = pd.melt(city_experience_ratings,
                    id_vars=['CITY'],
                    value_vars=experience_cols,
                    var_name='Experience Type',
                    value_name='Average Rating')

# 5. Create the grouped bar chart using Plotly Express
fig = px.bar(melted_df,
             x='CITY',
             y='Average Rating',
             color='Experience Type',
             barmode='group',
             title='Average Experience Ratings Across Top 10 Cities')

fig.update_layout(
    xaxis_title='City',
    yaxis_title='Average Rating',
    legend_title='Experience Type',
    xaxis={'categoryorder': 'total descending'}  # Sort x-axis by total count
)

fig.show()
```

```python
# @title Average approval rating
import plotly.express as px

def create_approval_rating_bar_chart(df, column, top_n=5, title=''):
    """
    Creates a bar chart for 'APPROVAL RATING' based on the top 'top_n' values of the specified 'column'.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column (str): The column to group by (e.g., 'STATE', 'CITY', 'AREA').
        top_n (int, optional): The number of top values to consider. Defaults to 5.
        title (str, optional): The title of the chart. Defaults to ''.

    Returns:
        plotly.graph_objs._figure.Figure: The Plotly figure object.
    """

    # 1. Get the top 'top_n' values based on frequency
    top_values = df[column].value_counts().head(top_n).index.tolist()

    # 2. Filter the dataframe for these top values
    filtered_df = df[df[column].isin(top_values)]

    # 3. Calculate average 'APPROVAL RATING' for each value
    approval_ratings = filtered_df.groupby(column)['APPROVAL RATING'].mean().reset_index(name='AVERAGE_APPROVAL_RATING')

    # 4. Create the bar chart using Plotly Express
    fig = px.bar(approval_ratings,
                 x=column,
                 y='AVERAGE_APPROVAL_RATING',
                 color='AVERAGE_APPROVAL_RATING',
                 title=title)

    fig.update_layout(
        xaxis_title=column,
        yaxis_title='Average Approval Rating',
        xaxis={'categoryorder': 'total descending'}  # Sort x-axis by total count
    )

    return fig


# Create and show the bar charts
fig_state = create_approval_rating_bar_chart(df, 'STATE', title='Average Approval Rating by Top 5 States')
fig_state.show()

fig_city = create_approval_rating_bar_chart(df, 'CITY', title='Average Approval Rating by Top 5 Cities')
fig_city.show()

fig_area = create_approval_rating_bar_chart(df, 'AREA', title='Average Approval Rating by Top 5 Areas')
fig_area.show()
```
